```python
# Import packages
import numpy as np 
import pandas as pd 
from scipy import signal
import matplotlib.pyplot as plt 
import math
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, precision_recall_fscore_support
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

## 1.0 Load Data

You will need to download the [DaLiAc database](https://mad-nas.cs.fau.de:8081/Research/ActivityNet/daliac.zip) and export all `dataset_x.txt` files into a folder named `dataset` within the same directory as this Jupyter Notebook.

The following code will load all datasets found within the `dataset` folder, the DaLiAc database contains 19 datasets (1 per person), therefore, only 19 dataset files are loaded (you may update this by modifying the range in the loop below). The first five rows and the last five rows of all combined datasets will be outputed after running this code.

```python
'''
Load data for 19 people from dataset folder (one dataset per person)
'''
def load_data():
    # Load all datasets
    for i in range(1,20):
        df = pd.read_csv('dataset/dataset_%d.txt' % i, sep=',', header=None)
        print('Loaded dataset_%d.txt (%d rows)' % (i, len(df)))
    
    # Print blank line to separate load status
    print('') 
    
    # Return loaded data
    return df


# Call our load_data() function
load_data()
```

## 2.0 Visualise Data

The following code will visualise the sensor data (accelerometer & gyroscope) collected from the device located on the 'Wrist' for all 13 activities performed by all 19 subjects within the DaLiAc database.

```python
'''
Visualise Accelerometer & Gyroscope sensor data from the wrist device for all activities
'''
def visualize_data():
    # Load all datasets
    df = load_data()
    
    # Plot data for all activities
    for activity in range(1, 14):
        df_values = df[df[24] == activity].values
    
        # Plot Accelerometer sensor data (Wrist only)
        fig = plt.figure()
        plt.plot(df_values[500:1500, 0:3])
        fig.suptitle('Activity %d - Wrist - Accelerometer' % activity, fontsize=20)
        plt.show()

        # Plot Gyroscope sensor data (Wrist only)
        fig = plt.figure()
        plt.plot(df_values[500:1500, 3:6])
        fig.suptitle('Activity %d - Wrist - Gyroscope' % activity, fontsize=20)
        plt.show()
        
        # Print blank line to separate each activity
        print('')


# Call our visualize_data() function
visualize_data()
```

## 3.0 Filter Data

The following code applies a 'lowpass' Butterworth filter to the sensor data (accelerometer & gyroscope) collected from the device located on the 'Wrist' for all 13 activities performed by all 19 subjects within the DaLiAc database in order to remove noise. Once the filter is applied the filtered data will be visualised in the output (you can compare this to the output of the code block above to see the change in noise).

```python
'''
Apply Butterworth filter to data from Accelerometer & Gryroscope sensors for sitting & lying activities
'''
def filter_data():
    # Load all datasets
    df = load_data()
    
    # Create Lowpass Butterworth filter for noise removal
    b, a = signal.butter(4, 0.04, 'low', analog=False)
    
    # Remove noise from all activity data using Lowpass Butterworth filter
    for activity in range(1, 14):
        df_values = df[df[24] == activity].values

        # Filter Accelerometer sensor data (Wrist only)
        for i in range(0, 3):
            df_values[:, i] = signal.lfilter(b, a, df_values[:, i])

        # Filter Gyroscope sensor data (Wrist only)
        for i in range(3, 6):
            df_values[:, i] = signal.lfilter(b, a, df_values[:, i])
    
        # Plot Accelerometer sensor data (Wrist only)
        fig = plt.figure()
        plt.plot(df_values[500:1500, 0:3])
        fig.suptitle('Activity %d - Wrist - Accelerometer' % activity, fontsize=20)
        plt.show()

        # Plot Gyroscope sensor data (Wrist only)
        fig = plt.figure()
        plt.plot(df_values[500:1500, 3:6])
        fig.suptitle('Activity %d - Wrist - Gyroscope' % activity, fontsize=20)
        plt.show()
        
        # Print blank line to separate each activity
        print('')
    

# Call our filter_data() function
filter_data()
```

## 4.0 Feature Engineering

The following code extracts features from the sensor data (accelerometer & gyroscope) collected from the device located on the 'Wrist' for all 13 activities performed by all 19 subjects within the DaLiAc database. The generated training & testing datasets (.csv files) will then be used with our Machine Learning models in '[5.0 Model Training & Evaluation](#5.0-Model-Training-&-Evaluation)'.

```python
'''
Extract features from raw data and return dataframe
'''
def extract_features(data, values, num):
    for s in range(num):
        if s < num - 1:
            sample_data = values[1000*s:1000*(s + 1), :]
        else:
            sample_data = values[1000*s:, :]
        
        # Extract min, max, mean, median, standard deviation & variance features
        feature_sample = []
        for i in range(24):
            feature_sample.append(np.min(sample_data[:, i]))
            feature_sample.append(np.max(sample_data[:, i]))
            feature_sample.append(np.mean(sample_data[:, i]))
            feature_sample.append(np.median(sample_data[:, i]))
            feature_sample.append(np.std(sample_data[:, i]))
            feature_sample.append(np.var(sample_data[:, i]))
        
        feature_sample.append(sample_data[0, -1])
        feature_sample = np.array([feature_sample])
        data = np.concatenate((data, feature_sample), axis=0)

    # Return dataframe
    return data


'''
Create testing & training datasets for use with our Machine Learning models
'''
def create_feature_datasets():
    training = np.empty(shape=(0, 145))
    testing = np.empty(shape=(0, 145))
    
    # Load all datasets
    for i in range(1, 20):
        df = pd.read_csv('dataset/dataset_%d.txt' % i, sep=',', header=None)
        print('Loaded dataset_%d.txt (%d rows)' % (i, len(df)))
        
        # Create Lowpass Butterworth filter for noise removal
        b, a = signal.butter(4, 0.04, 'low', analog=False)
        
        # Create features for all activities
        for activity in range(1, 14):
            df_values = df[df[24] == activity].values
            
            # Filter all sensor data
            for j in range(24):
                df_values[:, j] = signal.lfilter(b, a, df_values[:, j])
                
            training_length = math.floor(len(df_values) * 0.8)
            training_data = df_values[:training_length, :]
            testing_data = df_values[training_length:, :]

            # Segment data into training & testing data
            training = extract_features(training, training_data, (training_length // 1000 + 1))
            testing = extract_features(testing, testing_data, ((len(df_values) - training_length) // 1000 + 1))

    # Export training & testing data to .csv files
    df_training = pd.DataFrame(training)
    df_testing = pd.DataFrame(testing)
    df_training.to_csv('training_data.csv', index=None, header=None)
    print('Created training_data.csv')
    df_testing.to_csv('testing_data.csv', index=None, header=None)
    print('Created testing_data.csv')


# Call our create_feature_datasets() function
create_feature_datasets()
```

## 5.0 Model Training & Evaluation

The following code utilises the testing & training datasets (.csv files) generated in '[4.0 Feature Engineering](#4.0-Feature-Engineering)' with our KNN & SVM machine learning models. Once our classifier has been built with either model, we evaluate the performance of each model by outputting an accuracy score and a confusion matrix.

```python
'''
Use our testing & training data to build a classifier using either the KNN or SVM models; 
return confusion matrix to visualise model performance & accuracy
'''
def build_classifier(type):
    df_training = pd.read_csv('training_data.csv', header=None)
    df_testing = pd.read_csv('testing_data.csv', header=None)

    # Labels should start from 0 in sklearn
    y_train = (df_training[df_training.shape[1] - 1].values - 1).astype(int)
    X_train = df_training.drop([df_training.shape[1] - 1], axis=1).values
    print('Training input complete')
    
    y_test = (df_testing[df_testing.shape[1] - 1].values - 1).astype(int)
    X_test = df_testing.drop([df_testing.shape[1] - 1], axis=1).values
    print('Testing input complete')

    # Scale each feature by its maximum absolute value using the MaxAbsScaler feature normalisation method 
    scaler = preprocessing.MaxAbsScaler().fit(X_train)
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)
    print('Feature normlisation complete')
    
    if(type == 'KNN'):
        
        print('Running KNN classifier...')
        
        # Build KNN classifier
        knn = KNeighborsClassifier(n_neighbors=1)
        knn.fit(X_train, y_train)

        # Evaluate performance on testing set
        y_pred = knn.predict(X_test)
        
    elif(type == 'SVM'):
        
        print('Running SVM classifier...')
        
        # Build SVM classifier
        tuned_parameters = [{'kernel': ['rbf'], 
                             'gamma': [1e-1,1e-2, 1e-3, 1e-4],
                             'C': [1e-3, 1e-2, 1e-1, 1, 10, 100, 100]},
                            {'kernel': ['linear'], 
                             'C': [1e-3, 1e-2, 1e-1, 1, 10, 100]}]
        acc_scorer = make_scorer(accuracy_score)
        grid_obj  = GridSearchCV(SVC(), tuned_parameters, cv=10, scoring=acc_scorer).fit(X_train, y_train)
        clf = grid_obj.best_estimator_
        print('Best clf: ', clf)
        clf.fit(X_train, y_train)
        
        # Evaluate performance on testing set
        y_pred = clf.predict(X_test)
    
    # Print accuracy, precision, recall, F1 & confusion matrix
    print('Accuracy score: ', accuracy_score(y_test, y_pred))
    print('Precision score: ', precision_score(y_test, y_pred, average='macro'))
    print('Recall score: ', recall_score(y_test, y_pred, average='macro'))
    print('F1 score: ', f1_score(y_test, y_pred, average='macro'))
    print(confusion_matrix(y_test, y_pred))


# Build classifier with KNN model
build_classifier('KNN')

# Print blank line to separate each model
print('')

# Build classifier with SVM model
build_classifier('SVM')
```
